```python
import pandas as pd

data = pd.read_csv("metacritic_all_games_with_ign_reviews_corrected.csv")
data.head()
```

```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree

data['clean_review']= data['ign_review'].apply(lambda x:remove_punctuation(x))
data.head()
```

```python
data['review_lowerc']= data['clean_review'].apply(lambda x: x.lower())
data.head()
```

```python
from nltk import word_tokenize, sent_tokenize

data['review_token']= data['review_lowerc'].apply(lambda x: word_tokenize(x))
data.head()
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    output= [i for i in text if i not in stop_words]
    return output

data['review_stopwords']= data['review_token'].apply(lambda x:remove_stopwords(x))
```

```python
data.head()
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()

def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text
data['review_stemmed']=data['review_stopwords'].apply(lambda x: stemming(x))
```

```python
data.head()
```

```python
from nltk.stem import WordNetLemmatizer

wordnet_lemmatizer = WordNetLemmatizer()
```

```python
def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    return lemm_text
data['review_lemmatized']=data['review_stopwords'].apply(lambda x:lemmatizer(x))
```

```python
import numpy as np
index = data.index
data['random_number'] = np.random.randn(len(index))
train = data[data['random_number'] <= 0.8]
test = data[data['random_number'] > 0.8]
index
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from scipy import sparse


vectorizer = CountVectorizer(analyzer=lambda x: x)

X_train = sparse.hstack((vectorizer.fit_transform(train['review_lemmatized']),train[['meta_score']].values),format='csr')

X_test = sparse.hstack((vectorizer.transform(test['review_lemmatized']),test[['meta_score']].values),format='csr')

X2_columns=vectorizer.get_feature_names()+train[['review_lemmatized']].columns.tolist()

print(X_train.toarray())
print(X2_columns)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=10000)
```

```python
y_train = train['sentiment']
y_test = test['sentiment']
```

```python
lr.fit(X_train,y_train)
```

```python
predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
new = np.asarray(y_test)
confusion_matrix(predictions,y_test)
```

```python
print(classification_report(predictions,y_test))
```
